```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# Seting seed
random.seed(42)
np.random.seed(42)
```

```python
# Customer count
num_customers = 10000
```

```python
customers = pd.DataFrame({"customer_id": range(1, num_customers + 1), "age": np.random.randint(18, 70, num_customers),"country": np.random.choice(["India", "USA", "UK", "Germany"], num_customers),
                          "account_age_days": np.random.randint(1, 3650, num_customers)})
```

```python
customers.head()
```

```python
# Devices and Ips
num_devices = 5000
num_ips = 8000
```

```python
# device_1, ip_1
devices = [f"device_{i}" for i in range(num_devices)]
ips = [f"ip_{i}" for i in range(num_ips)]
```

```python
# Transactions
num_transactions = 200000
```

```python
# Swipe, Internet, cash
transaction_types = ["POS", "Online", "ATM"]
```

```python
merchant_categories = ["Food", "Electronics", "Travel", "Shopping"]
```

```python
# Making a table:
transactions = pd.DataFrame({
        "transaction_id": range(1, num_transactions + 1),
        "customer_id": np.random.choice(customers["customer_id"], num_transactions),
        "amount": np.round(np.random.exponential(scale=100, size=num_transactions) + 1,2),
        "transaction_type": np.random.choice(transaction_types, num_transactions),
        "merchant_category": np.random.choice(merchant_categories, num_transactions),
})
```

```python
# Time stamp
start_date = datetime(2023, 1, 1)
transactions["timestamp"] = [start_date + timedelta(minutes=random.randint(0, 60*24*365))
                            for _ in range(num_transactions)
                            ]
```

```python
# Merging both tables
data = transactions.merge(customers, on="customer_id", how="left")
```

```python
# Behaviours
data["device_id"] = np.random.choice(devices, len(data)) #Assign device to each transaction, 
data["ip_id"] = np.random.choice(ips, len(data)) # What device type
data["ip_country"] = np.random.choice(["India", "USA", "UK", "Germany"], len(data)) # Where transaction came from
data["hour"] = data["timestamp"].dt.hour #Extract hour (0–23)
data["transactions_last_hour"] = np.random.randint(1, 10, len(data)) #How many transactions recently
```

```python
# Fraud Ring.
fraud_devices = random.sample(devices, 50)


data["is_fraud_ring"] = data["device_id"].apply(
    lambda x: 1 if x in fraud_devices else 0
)

```

```python
# Fraud fucntion
def detect_fraud(row):
    score = 0

    if row["amount"] > 1500 and row["account_age_days"] < 60:
        score += 1

    if row["transactions_last_hour"] > 7:
        score += 1

    if row["country"] != row["ip_country"]:
        score += 1

    if row["hour"] < 4:
        score += 1

    if row["is_fraud_ring"] == 1:
        score += 2  # strong signal

    
    if score >= 3:
        fraud = 1
    elif score == 2 and random.random() < 0.5:
        fraud = 1
    else:
        fraud = 0

    return fraud

    
```

```python
# Fraud Label
data["fraud"] = data.apply(detect_fraud, axis=1)

#Distribution
data["fraud"].value_counts(normalize=True)
```

```python
#Edge List
edges = []

# Customer → Transaction
edges += list(zip(
    "cust_" + data["customer_id"].astype(str),
    "txn_" + data["transaction_id"].astype(str)
))

# Transaction → Device
edges += list(zip(
    "txn_" + data["transaction_id"].astype(str),
    data["device_id"]
))

# Transaction → IP
edges += list(zip(
    "txn_" + data["transaction_id"].astype(str),
    data["ip_id"]
))
```

```python
edges_df = pd.DataFrame(edges, columns=["source", "target"])

edges_df.head()
```

```python
data.to_csv("gnn_fraud_dataset.csv", index=False)
edges_df.to_csv("gnn_edges.csv", index=False)
```
